```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
np.random.seed(0)
n_samples = 250
# Вариант 7
X = np.random.rand(n_samples, 2)
Y = (((3*X[:, 0]-0.3))+(-0.5*X[:, 1]-0.3) < 0.5)
w = np.array([0,0])
b = 0
```

```python
def ADALINE(X,Y,n_samples,W,b,alpha):
    plt.scatter(X[Y,0],X[Y,1])
    plt.scatter(X[~Y,0],X[~Y,1],color = 'r')
    targ = []
    epochs =0
    MSE = 0
    for y in Y:
        if y == True:
            targ.append(1)
        else:
            targ.append(-1)
    while(epochs<10):
        for i in range(X.shape[0]):
            y_s = np.dot(W,X[i])+b
            error1 =(targ[i]-y_s)**2
            error =2*(targ[i]-y_s)

            MSE+=error1
            W = W + alpha*error *X[i]
            b = b + alpha*error
        epochs+=1
        if(MSE<0.1):

            break

    X = np.array([0,1])
    Y = -((W[0]/W[1]))*X - b/W[1]
    print("W = ", W )
    print("b = ", b)
    #print("MSE = ", MSE)
    plt.plot(X,Y)
    plt.grid()


alpha = 0.05
ADALINE(X,Y,n_samples,w,b,alpha)
```
