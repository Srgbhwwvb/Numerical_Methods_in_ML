```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
n_samples = 250
# Вариант 7
X = np.random.rand(n_samples, 2)
Y = (((3*X[:, 0]-0.3))+(-0.5*X[:, 1]-0.3) < 0.5)
```

```python
def perceptron(n,w,b):
  i=0
  c=0
  iter=0
  while True:
    y=int((w[0]*X[i][0]+w[1]*X[i][1] + b)>=0)
    e=Y[i]-y
    w=w+e*X[i]
    b=b+e
    if(e==0):
      c+=1
    else:
      c=0
    i+=1
    iter+=1
    if(i>=n):
      i=0
    if(c>=n):
      break

  print(w)
  print(b)
  print("Количество итераций: ",iter)
  return w,b

w=np.array([0, 0])
b=0.0
w,b=perceptron(n_samples,w,b)

plt.axes(xlim=(0, 1), ylim=(0, 1))
plt.scatter(X[Y,0],X[Y,1])
plt.scatter(X[~Y,0], X[~Y,1],c="red")
x=np.arange(-10,10, 1)
plt.plot(x,-b/w[1]-x*w[0]/w[1])
```
